```python
# Convert dataset into DataFrame
import pandas as pd
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df['species'] = iris.target
```

```python
df['species'] = df['species'].map({
    0: 'setosa',
    1: 'versicolor',
    2: 'virginica'
})
```
